Next, I will inspect the correlationship of features with the target.

Pearson correlation will be used, pearson correlation can be used for numerical and ordinal columns(with mapping if applicable), but it cannot be used for nominal categorical variables. 

It is because Pearson correlation measures linear relationships, it assumes the numbers have order and distance, so it can calculate if and how an increase in feature x would affect target y. For nominal categories, numeric labels have no logical meaning, so it cannot be used 

However, in this dataset, not all ordinal columns are numbers, there are ratings from 1 to 10, and there are also string categories like “Ex", "Gd", "TA", for Excellent, Good, and Average, for such columns, calculating the Pearson correlation coefficient will not work, encoding must be done before correlating

I will first map catgories to ordered numbers

Among the string ordinal columns, Some attributes have structured missingness, where NaN means "no such amenities", but it will be worthwhile to check if any column has actual missing data

```python
ordinal_cols = ['OverallQual','OverallCond','ExterQual', 'ExterCond',
                'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
                'HeatingQC','KitchenQual','FireplaceQu','GarageQual',
                'GarageCond','PoolQC'
]

# check with data_description.txt to see which are the columns without structured missingness 
no_structured_missing = ['OverallQual','OverallCond','ExterQual', 'ExterCond',
                         'HeatingQC','KitchenQual']

structured_missing = [col for col in ordinal_cols if col not in no_structured_missing]
print(structured_missing)

# check for missing data 
for col in no_structured_missing:
  if data[col].isna().any():
    print(col)
```

There is no true missingness found in ordinal columns, I can start encoding

For columns with structured missingness, for example if a house does not even have a Pool, then it is of course worse than a house with a poor pool, I will therefore treat "no such amenity" as being lower tier than "Poor amentity" 

```python
data_ord_encode = data.copy() # create a copy of data

for col in structured_missing:
  data_ord_encode[col] = data_ord_encode[col].fillna("None")
```

```python
data_ord_encode.head(15)
```

Each string ordinal categorical attributes have different names for the tiers, for example from data_description.txt: 

BsmtExposure: Refers to walkout or garden level walls

       Gd	Good Exposure
       Av	Average Exposure (split levels or foyers typically score average or above)	
       Mn	Mimimum Exposure
       No	No Exposure
       NA	No Basement
	
BsmtFinType1: Rating of basement finished area

       GLQ	Good Living Quarters
       ALQ	Average Living Quarters
       BLQ	Below Average Living Quarters	
       Rec	Average Rec Room
       LwQ	Low Quality
       Unf	Unfinshed
       NA	No Basement

In my ordinal_cols, the attributes can be classified to have 4 types of different tier systems, below I will encode them according to the system they belong 

```python
# Previously, I have already filled the NaNs of ordinal attributes with None, where applicable 
Qual_and_Con_scale = ['None','Po','Fa','TA','Gd','Ex']
Exposure_scale = ['None','No','Mn','Av','Gd']
Fintype_scale = ['None','Unf','LwQ','Rec','BLQ','ALQ','GLQ']


# define the columns that uses each scale 
qnc_cols = ['ExterQual','ExterCond','BsmtQual','BsmtCond','HeatingQC','KitchenQual','FireplaceQu',
'GarageQual','GarageCond','PoolQC']
exp_col = ['BsmtExposure']
fintype_cols = ['BsmtFinType1','BsmtFinType2']


# Generate mappings
qnc_map = {k: i for i, k in enumerate(Qual_and_Con_scale)}
exp_map = {k: i for i, k in enumerate(Exposure_scale)}
fintype_map = {k: i for i, k in enumerate(Fintype_scale)}


# apply the mappings
for col in qnc_cols:
  data_ord_encode[col] = data_ord_encode[col].map(qnc_map)

for col in exp_col:
  data_ord_encode[col] = data_ord_encode[col].map(exp_map)

for col in fintype_cols:
  data_ord_encode[col] = data_ord_encode[col].map(fintype_map)

```

Now we inspect the correlation between ordinal categorical features and the target, also that between numerical features and the target

```python
ordinal_corr = (
  data_ord_encode[ordinal_cols + ['SalePrice']]
  .corr()['SalePrice']
  .drop('SalePrice')
  .sort_values(ascending=False)
)

print(ordinal_corr)
```
